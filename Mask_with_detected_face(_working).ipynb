```python
import os
import cv2
import shutil

# Paths for input folder and output folder
input_folder = '/content/drive/MyDrive/dataset/test/'  # Replace with the path to your input image folder
output_folder = '/content/test_df/'  # Replace with the path to your output image folder

# Ensure the output folder exists, create it if necessary
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Load the pre-trained Haarcascades face detection classifier
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Process each image in the input folder
for class_folder in os.listdir(input_folder):
    if os.path.isdir(os.path.join(input_folder, class_folder)):
        class_path = os.path.join(input_folder, class_folder)

        for filename in os.listdir(class_path):
            if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif')):
                image_path = os.path.join(class_path, filename)

                print(f"Processing image: {image_path}")

                # Load the captured photo
                captured_image = cv2.imread(image_path)

                if captured_image is None:
                    print(f"Error: Could not load image: {image_path}")
                    continue

                # Convert the image to grayscale
                gray_image = cv2.cvtColor(captured_image, cv2.COLOR_BGR2GRAY)

                # Detect faces in the grayscale image
                faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

                # If faces are detected, proceed to crop and save the face region
                if len(faces) > 0:
                    x, y, w, h = faces[0]  # Assuming only one face is detected, you can modify this accordingly
                    face_region = captured_image[y:y+h, x:x+w]  # Crop the face region from the color image

                    # Organize output images into subfolders based on their classes
                    output_class_folder = os.path.join(output_folder, class_folder)
                    if not os.path.exists(output_class_folder):
                        os.makedirs(output_class_folder)

                    # Save the detected face region image in color
                    output_path = os.path.join(output_class_folder, f'detected_face_{filename}')
                    cv2.imwrite(output_path, face_region)
                    print(f"Detected face region image saved at: {output_path}")
                else:
                    print(f"No faces detected in {filename}.")
```

```python
from PIL import Image
import colorsys
import numpy as np
import os
import shutil

# Define the input and output folders
input_folder = '/content/test_df'  # Replace with the path to your input image folder
output_folder = '/content/test/'  # Replace with the path to your output image folder (use 'test')

# Define the threshold values for each color space
# Adjust these threshold values according to your requirements
rgb_threshold = (100, 100, 100)  # RGB thresholds (R, G, B)
hsv_threshold = (0.2, 0.5, 0.5)  # HSV thresholds (H, S, V)
ycbcr_threshold = (100, 128, 128)  # YCbCr thresholds (Y, Cb, Cr)

# Process each image in the input folder and its subfolders
for root, dirs, files in os.walk(input_folder):
    for filename in files:
        if filename.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif')):
            image_path = os.path.join(root, filename)

            # Load the image using PIL
            image = Image.open(image_path)

            # Convert the image to a NumPy array
            pixel_array = np.array(image)

            # Create an empty mask with the same shape as the image
            mask = np.zeros_like(pixel_array)

            # Iterate through each pixel in the image
            for i in range(pixel_array.shape[0]):
                for j in range(pixel_array.shape[1]):
                    # Get the RGB values of the current pixel
                    r, g, b = pixel_array[i, j]

                    # Convert RGB to HSV
                    h, s, v = colorsys.rgb_to_hsv(r / 255.0, g / 255.0, b / 255.0)

                    # Convert RGB to YCbCr
                    y = 0.299 * r + 0.587 * g + 0.114 * b
                    cb = (b - y) * 0.564 + 128
                    cr = (r - y) * 0.713 + 128

                    # Check if the pixel values exceed the threshold values in each color space
                    rgb_condition = all(value > threshold for value, threshold in zip((r, g, b), rgb_threshold))
                    hsv_condition = all(value > threshold for value, threshold in zip((h, s, v), hsv_threshold))
                    ycbcr_condition = all(value > threshold for value, threshold in zip((y, cb, cr), ycbcr_threshold))

                    # Decide whether to mask the pixel or not based on the threshold conditions
                    if rgb_condition or hsv_condition or ycbcr_condition:
                        mask[i, j] = pixel_array[i, j]

            # Convert the NumPy mask array back to an image
            mask_image = Image.fromarray(mask)

            # Get the subfolder name (e.g., 'white', 'brown', 'black')
            subfolder_name = os.path.basename(root)

            # Create subfolders in the output directory if they don't exist
            output_subfolder = os.path.join(output_folder, subfolder_name)
            if not os.path.exists(output_subfolder):
                os.makedirs(output_subfolder)

            # Save the masked image with non-skin pixels in the output folder
            output_path = os.path.join(output_subfolder, filename)
            mask_image.save(output_path)
            print(f"Masked image saved at: {output_path}")

print("All masked images saved in corresponding subfolders.")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set the path to your dataset directory
dataset_dir = '/content/train'

# Define the parameters
batch_size = 16
epochs = 50
image_size = (128, 128)

# Create an ImageDataGenerator
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.3
)

train_generator = datagen.flow_from_directory(
    dataset_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    dataset_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Build the model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(len(train_generator.class_indices), activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(train_generator, epochs=epochs, validation_data=validation_generator)

# Save the model
model.save('face_classifier.h5')
```

```python
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Load the trained model
model = tf.keras.models.load_model('face_classifier.h5')

# Set the path to your test dataset directory
test_dataset_dir = '/content/test'

# Create a test ImageDataGenerator
test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    test_dataset_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Predict classes for test images
predictions = model.predict(test_generator)
predicted_classes = predictions.argmax(axis=-1)
true_classes = test_generator.classes

# Calculate the confusion matrix
confusion = confusion_matrix(true_classes, predicted_classes)


# Plot the confusion matrix using Seaborn
plt.figure(figsize=(8, 6))
# Define a colormap that includes a "skin" color scheme
custom_cmap = sns.color_palette("YlOrRd")  # You can choose a different colormap

# Create a heatmap with the custom colormap
sns.heatmap(confusion, annot=True, fmt='d', cmap=custom_cmap, xticklabels=train_generator.class_indices.keys(), yticklabels=train_generator.class_indices.keys())
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')

# Save the confusion matrix plot as an image
confusion_matrix_image_path = 'confusion_matrix.png'
plt.savefig(confusion_matrix_image_path)

# Generate classification report
report = classification_report(true_classes, predicted_classes, target_names=train_generator.class_indices.keys())
print(report)
```
